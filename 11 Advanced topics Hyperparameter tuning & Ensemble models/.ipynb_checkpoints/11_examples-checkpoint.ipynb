### 1.1 Support Vector Classifier

```python
sss=model_selection.StratifiedShuffleSplit(n_splits=5,test_size=0.1) #分層抽樣
# 先依照Y分成三組(0,1,2，一組總共50筆資料)，其中0.1拿來當作testing data (5筆)。最後每個Y都有45筆sample, 5筆test
# 這件事做5次 (n_split=5)
len(list(sss.split(X,Y))) # 這個會回傳要拿來當sample還是test的資料的index

dataSplit = np.array([['sample1_index','test1_index'],['sample2_index','test2_index']]) # 結構長這樣
for sample_index, test_index in dataSplit:
    print(sample_index, test_index)

for train_index, test_index in sss.split(X, Y): # 這個迴圈是五圈
    print('Training-Testing overlaps:',set(train_index).intersection(test_index))
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    print("Y_train:",sum(Y_train==0),sum(Y_train==1),sum(Y_train==2),"Y_test",sum(Y_test==0),sum(Y_test==1),sum(Y_test==2),'\n')
```

```python
model=svm.SVC(C=10)
acc=[]
# 5 realizations from Binomial(15,p)
for train_index, test_index in sss.split(X, Y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    model.fit(X_train[:,0:2],Y_train) #training
    acc.append(sum(model.predict(X_test[:,0:2])==Y_test)) #testing, SVC is expecting only 2 features as input.
print(acc) # 5個array, 每個裡面都是15個testing data 正確與否
print(np.mean(acc)/15) # Xbar (N = 5) estimator of p in Binomial(15, p)
```

### 1.2 Grid Search (Exhaustive Search)

```python
C=[10**i for i in range(-7,7)]
print(C)
```

```python
import time
xx=[]
yy=[]
t0=time.time()
for i in range(-6,6):
    print(i,end=' ')
    xx.append(10**i)
    model=svm.SVC(C=xx[-1]) # tuning the hyperparameter C
    acc=[]
    for train_index, test_index in sss.split(X, Y):
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        model.fit(X_train[:,0:2],Y_train) #training
        acc.append(np.mean(model.predict(X_test[:,0:2])==Y_test)) #testing
    yy.append(np.mean(acc))
print('Search time:', time.time()-t0)
print(yy)
plot(xx,yy,'o-');
```

```python
model=svm.SVC()
parameters={'C':[10**i for i in range(-6,6)]}
clf=model_selection.GridSearchCV(model, parameters, cv=5)
clf.fit(X[:,0:2], Y)
print('C:',clf.cv_results_['param_C'])
print('Mean Test Score:',clf.cv_results_['mean_test_score'])
print('Best C:',clf.best_params_)
```

### 1.3 HyperOpt

```python
# !pip install git+https://github.com/hyperopt/hyperopt.git
from hyperopt import fmin, tpe, rand, hp

def F(x):
    return(x**2)

best_x = fmin(fn=F,
    space=hp.uniform('x', -6, 6),
    algo=rand.suggest,
    max_evals=100)
print(best_x)
```

```python
import time

def F(x):
    model=svm.SVC(C=10**x) # tuning the hyperparameter C
    acc=[]
    for train_index, test_index in sss.split(X, Y):
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        model.fit(X_train[:,0:2],Y_train) #training
        acc.append(np.mean(model.predict(X_test[:,0:2])==Y_test)) #testing
    return(1-np.mean(acc)) # error

t0=time.time()
best_x = fmin(fn=F,
    space=hp.uniform('x', -5, 5),
    algo=tpe.suggest, # try rand.suggest
    max_evals=100)
print('Search Time:',time.time()-t0)
print(best_x)
```

## 2 Ensemble Learning

### 2.0 Decision Tree

```python
skf=model_selection.StratifiedKFold(n_splits=5,shuffle=True)
for train_index, test_index in skf.split(X, Y):
    print('Training-Testing overlaps:',set(train_index).intersection(test_index))
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    print("Y_train:",sum(Y_train==0),sum(Y_train==1),sum(Y_train==2),"Y_test",sum(Y_test==0),sum(Y_test==1),sum(Y_test==2),'\n')
```

```python
model=tree.DecisionTreeClassifier()
acc=[]
for train_index, test_index in skf.split(X, Y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    model.fit(X_train[:,0:2],Y_train) #training
    acc.append(model.predict(X_test[:,0:2])==Y_test) #testing
print(np.mean(acc))
```

### 2.1 Random Forest

```python
skf=model_selection.StratifiedKFold(n_splits=5,shuffle=True)
model=ensemble.RandomForestClassifier(n_estimators=100, max_depth=2) # 給 1000個小學生做微積分題目正確率也不會提高，給大學生做可能會做但是每個人的意見差很多，所以需要多點人
acc=[]
for train_index, test_index in skf.split(X, Y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    model.fit(X_train[:,0:2],Y_train) #training
    acc.append(model.predict(X_test[:,0:2])==Y_test) #testing
print(np.mean(acc))
```

```python
# !pip install git+https://github.com/thuijskens/scikit-hyperband
from hyperband import HyperbandSearchCV
from scipy.stats import randint as sp_randint
model=ensemble.RandomForestClassifier()

param_dist = {
    'max_depth': [3, None],
    'max_features': sp_randint(1, 3),
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}

clf = HyperbandSearchCV(model, param_dist, resource_param='n_estimators',min_iter=1, max_iter=100)
clf.fit(X[:,0:2],Y)

print(clf.best_params_)
print(clf.best_score_)
```

### 2.2 Bagging (Bootstrap Aggregating)

```python
model=ensemble.BaggingClassifier(tree.DecisionTreeClassifier(max_depth=1),max_samples=1.0, max_features=1.0, n_estimators=100)
#model=ensemble.BaggingClassifier(svm.SVC(0.1),max_samples=1.0, max_features=1.0, n_estimators=100)


acc=[]
for train_index, test_index in skf.split(X, Y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    model.fit(X_train[:,0:2],Y_train) #training
    acc.append(model.predict(X_test[:,0:2])==Y_test) #testing
print(np.mean(acc))
```

### 2.3 Boosting

#### 2.3.1 AdaBoost

```python
#model=ensemble.AdaBoostClassifier(svm.SVC(0.1),n_estimators=100,algorithm="SAMME")
model=ensemble.AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=1),n_estimators=100,algorithm="SAMME")

acc=[]
for train_index, test_index in skf.split(X, Y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    model.fit(X_train[:,0:2],Y_train) #training
    acc.append(model.predict(X_test[:,0:2])==Y_test) #testing
print(np.mean(acc))
```
