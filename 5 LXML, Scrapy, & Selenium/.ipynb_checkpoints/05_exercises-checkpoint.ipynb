## 進一步搜尋 Boy-Girl 版資訊 (8 points)

```python
import urllib, lxml.html
URL='http://www.ptt.cc'
URN='/bbs/Boy-Girl/'
h={'User-Agent':'Mozilla/5.0'}
```

# 1. index.html 右上角 [< 上頁 ] 中包含了總頁數資訊，請用 LXML 抓出此經常變動的數字。(2 points)

```python
# Plase write your codes here...
r=urllib.request.Request(URL+URN,headers=h)
data=urllib.request.urlopen(r).read()
t=lxml.html.fromstring(data.decode('utf-8')) 
button=t.xpath('//a[@class="btn wide"]') # 找到 a裡面有 class是"btn wide"的部分
for text in button: # 在x這個List裡面找到他的內文符合'< 上頁'的component
    if text.xpath('text()')[0] == '‹ 上頁':
        Pagehref = text.xpath("@href")[0] #將該 component的 href存起來

print(Pagehref) # "上頁"按鈕的連結
```

```python
# 在網址中定位頁碼
def findPage(href):
    startIndex = href.find('index') + 5
    endIndex = href.find('.html')
    newestPage = int(href[startIndex:endIndex]) + 1 #由於是找上頁按鈕的連結，總頁數要再 + 1
    return  newestPage
    
newestPage = findPage(Pagehref)
print(newestPage)
```

# 2. 請用 LXML 找出距離現在時間最近的一篇[爆]文標題與 URN (有可能需要翻頁) 。(3 points)

```python
# Plase write your codes here...
push = ""
title = ""
targetURN = ""
Page = findPage(Pagehref)

while push != "爆":
    URN=f'/bbs/Boy-Girl/index{Page}.html'
    r=urllib.request.Request(URL+URN,headers=h)
    data=urllib.request.urlopen(r).read()
    t=lxml.html.fromstring(data.decode('utf-8')) 
    z=t.xpath('//div[@class="r-ent"]')

    for post in z:
        try:
            push = post.xpath('./div[@class="nrec"]//text()')[0]
        except:
            push = "0"
        try:
            title =  post.xpath('./div[@class="title"]/a/text()')[0]
        except:
            title = "deleted"
        if push == "爆":
            targetURN = post.xpath('./div[@class="title"]/a/@href')[0]
            break
    
    Page-=1

    
print(f'{push} {title} {targetURN}')
```

# 3. 請用 Selenium 在 index.html 往前翻三頁，每頁拍一張照片，在notebook內顯示。(3 points)

```python
# Plase write your codes here...
from selenium import webdriver

Page = findPage(Pagehref)
URL='https://www.ptt.cc/'
URN=f'/bbs/Boy-Girl/index{Page}.html'
driver=webdriver.Edge() # try Firefox() or Edge()
ScreenshotList = []

for i in range(3):
    URN=f'/bbs/Boy-Girl/index{Page}.html'
    driver.get(URL+URN)
    ScreenshotList.append(f'Page{Page} Screenshot.png')
    driver.save_screenshot(ScreenshotList[i])
    Page-=1
```
